```python
#remove the redundant data which is no longer of use
student=data.drop(['race/ethnicity','parental level of education'],axis=1)
student.head()
```

```python
#lloking for outliers
#outlier is simply a data which is different from other observations
#relationship analysis
```

```python
#correlation matrix- each cell shows correlation b/w two variables
#use to summarize data as an input for more advanced analysis
corelation=student.corr()
```

```python
#now putting this inside heatmap
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
#pairplot
sns.pairplot(student)
```

```python
#scatterplot
#relationplot
sns.relplot(x='math score',y='reading score',hue='gender',data=student)
```
